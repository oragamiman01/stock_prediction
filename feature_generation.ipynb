```python
from openbb import obb
import matplotlib.pyplot as plt
obb.technical
stock_data = obb.equity.price.historical(symbol="TSLA", start_date="2023-01-01", provider="yfinance")
```

```python
df = stock_data.to_df().reset_index(level=['date'])
df = df.drop(columns=['dividends', 'stock_splits'])
feature = obb.technical.rsi(data=df, target='close')
df = feature.to_df().reset_index(level=['date'])
df
```

```python
df.plot(x='date', y=['close_RSI_14'], figsize=(8,3))
# df.plot(x='date', y='close', figsize=(10,5))
plt.show()
```
